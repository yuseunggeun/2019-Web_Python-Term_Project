### 4.1.1)플레이어의 움직임 만들기 

Player 클래스에 점프와 움직임을 갱신하는 함수를 추가했다.

        global hit
        self.rect.y += 1
        if not hit:
            hit = pygame.sprite.collide_rect(self,platform)

def update(self): #플레이어의 움직임 갱신
    global hit
    self.acc = 1.3 #중력가속도를 의미한다.
    keys = pygame.key.get_pressed() #키보드가 눌렸을때 무슨키가 눌렸는지 판단한다.
    if keys[pygame.K_LEFT]:  #왼쪽 방향키를 눌렀을 때 플레이어의 좌표를 왼쪽으로 이동시킨다. 
        self.rect.x -= self.speed
        if self.rect.x <0: #화면밖을 벗어나지 못하게 만듦
            self.rect.x = 0
        self.direction = 'left' #이후 bullet의 이동방향 결정에 필요한 변수
    if keys[pygame.K_RIGHT]: #오른쪽 방향키를 눌렀을 때 플레이어의 좌표를 왼쪽으로 이동시킨다. 
        self.rect.x += self.speed
        if self.rect.x >750: #화면밖을 벗어나지 못하게 만듦
            self.rect.x = 750
        self.direction = 'right'
    if hit == True:
        if self.rect.y > 530:
            self.rect.y = 530
            self.jumpspeed = 20
            hit = 0
        else:
            self.jumpspeed -= self.acc
            self.rect.y -= self.jumpspeed

````python
#'!pip install pygame'을 실행하여 주피터노트북 내에 pygame 모듈을 설치해야 실행됩니다.
import pygame
import os 
import sys
import time
import random

class Player(pygame.sprite.Sprite): #플레이어 클래스 정의

    def __init__(self): #기본위치,이미지,속도 정의
        pygame.sprite.Sprite.__init__(self) #스프라이트화 시킴
        self.speed = 7
        self.jumpspeed = 20
        self.direction = 'right'
        self.image = pygame.image.load('a.png')
        self.image = pygame.transform.scale(self.image,(50,50))
        self.rect = self.image.get_rect()
        self.rect.x = 0
        self.rect.y = 530

    def jump(self): #땅과 충돌여부 판단
        global hit
        self.rect.y += 1
        if not hit:
            hit = pygame.sprite.collide_rect(self,platform)
        self.rect.y -= 1
        
    def update(self): #플레이어의 움직임 갱신
        global hit
        self.acc = 1.3
        keys = pygame.key.get_pressed()
        if keys[pygame.K_LEFT]:
            self.rect.x -= self.speed
            if self.rect.x <0:
                self.rect.x = 0
            self.direction = 'left'
        if keys[pygame.K_RIGHT]:
            self.rect.x += self.speed
            if self.rect.x >750:
                self.rect.x = 750
            self.direction = 'right'
        if hit == True:
            if self.rect.y > 530:
                self.rect.y = 530
                self.jumpspeed = 20
                hit = 0
            else:
                self.jumpspeed -= self.acc
                self.rect.y -= self.jumpspeed

    
        
class Platform(pygame.sprite.Sprite): #땅 클래스 정의
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.image = pygame.image.load('ground.png')
        self.rect = self.image.get_rect()
        self.rect.x = 0
        self.rect.y = 580
       
    
class Bullet(pygame.sprite.Sprite): #총알 클래스 정의
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.image = pygame.image.load('attack2.png')
        self.image = pygame.transform.scale(self.image,(25,25))
        self.rect = self.image.get_rect()
        if nemo.direction == 'left': #왼쪽 방향이면 x값이 작아지게
            self.direction = -15
        elif nemo.direction == 'right': #오른쪽 방향이면 x값이 커지게
            self.direction = +15
            
    def update(self): #총알 움직임 갱신
        self.rect.x += self.direction
  
                   
class Enemy(pygame.sprite.Sprite): #적 클래스 정의
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.image = pygame.image.load('enemy.png')
        self.image = pygame.transform.scale(self.image,(50,100))
        self.rect = self.image.get_rect()
        self.hp = 300
        self.rect.x = 500
        self.rect.y = 480

        
class Hit_effect(pygame.sprite.Sprite): #타격효과 클래스 정의
    def __init__(self):
        pygame.sprite.Sprite.__init__(self)
        self.image = pygame.image.load('hit_effect1.png')
        self.image = pygame.transform.scale(self.image,(50,50))
        self.rect = self.image.get_rect()


#------------------------game prepare---------------------------------------------

pygame.init()
screen = pygame.display.set_mode((800,600)) #창 크기 설정
pygame.display.set_caption('game')
clock = pygame.time.Clock()

nemo = Player() #이하 nemo=플레이어
platform = Platform() # platform =땅
enemy = Enemy() #enemy = 적

all_list = pygame.sprite.Group() #모든 스프라이트 객체 저장
bullet_list = pygame.sprite.Group() #총알 객체 저장
enemy_list = pygame.sprite.Group() #적 객체 저장
all_list.add(nemo) 
enemy_list.add(enemy)
all_list.add(enemy)
all_list.add(platform)


effect_list = pygame.sprite.Group() #효과 객체 저장

hit = 0

previous_time = pygame.time.get_ticks() #총알 딜레이 목적
#--------------------------------game start----------------------------------
while True:
    screen.fill(pygame.color.Color(255,255,255)) #하얀색 배경
    for event in pygame.event.get(): #파이게임 종료
        if event.type == pygame.QUIT:
            pygame.quit()
            sys.exit()
    key_event = pygame.key.get_pressed() #위방향키가 눌리면 점프
    if key_event[pygame.K_UP]:
        nemo.jump()
        
    if key_event[pygame.K_z]: #z키 누르면 총알 발사
        current_time = pygame.time.get_ticks()
        if current_time - previous_time > 100: # 이전 총알 발사후 0.1초 지났으면
            previous_time = current_time
            bullet = Bullet() #총알 객체 생성
            bullet.rect.center = nemo.rect.center
            all_list.add(bullet)
            bullet_list.add(bullet)
    
    all_list.update() #모든 객체 움직임 갱신
    
    for bullet in bullet_list: #총알과 적 충돌 판정
        hit_list = pygame.sprite.spritecollide(bullet,enemy_list,False)
        
        for enemy in hit_list:
            bullet_list.remove(bullet) #적과 부딪힌 총알 제거
            all_list.remove(bullet)
            enemy.hp -= 1 #적 hp 감소
            
            hit_effect1 = Hit_effect() #타격효과 생성
            hit_effect1.rect.center = enemy.rect.center
            angle = random.randint(0,360) #타격효과 무작위 회전
            hit_effect1.image = pygame.transform.rotate(hit_effect1.image,angle)
            effect_list.add(hit_effect1)
            all_list.add(hit_effect1)
            previous_time1 = pygame.time.get_ticks()
            
        if bullet.rect.x < 0 or bullet.rect.x > 800: #총알이 화면을 나가면
            bullet_list.remove(bullet)
            all_list.remove(bullet)
            
    for effect in effect_list:
        current_time1 = pygame.time.get_ticks()
        if current_time1 - previous_time1 > 60: #효과 보여주고 0.06초 지났으면
            previous_time1 = current_time1
            effect_list.remove(effect)
            all_list.remove(effect) 
        else:
            if len(effect_list) > 1: #이전에 보여준 효과가 겹치면 제거
                effect_list.remove(effect)
                all_list.remove(effect)
 
    all_list.draw(screen) #모든 객체 화면에 표시
    
    
    
    enemy_hp_back = pygame.image.load('enemy_hp_back.png') #상단에 적 hp바 표시
    enemy_hp_back = pygame.transform.scale(enemy_hp_back,(800,50))
    screen.blit(enemy_hp_back,(0,0))
    enemy_hp = pygame.image.load('enemy_hp.png')
    if 787-(300-enemy.hp)*787/300 <= 0:
        enemy_hp = pygame.transform.scale(enemy_hp,(0,42))
    else:
        enemy_hp = pygame.transform.scale(enemy_hp,(int(787-(300-enemy.hp)*787/300),42))
    
    screen.blit(enemy_hp,(7,5))
    
    pygame.display.update()
    clock.tick(60)
````
